# **Menghitung Indikator Regional**

```python
sns.countplot(hd_2020['Regional indicator'],color="green", facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("BrBG", 2));
plt.xticks(rotation=60)
plt.show()
```

#**Distribusi Fitur**

```python
# Feature Set-3
list_features=['Social support', 'Freedom to make life choices','Generosity', 'Perceptions of corruption']
sns.boxplot(data=hd_2020.loc[:,list_features],
    notch=True, showcaps=False,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "orange"})
plt.show()
```

```python
# Feature Set-2
list_features=['Ladder score', 'Logged GDP per capita']
sns.boxplot(data=hd_2020.loc[:,list_features],
    notch=True, showcaps=False,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "yellow"})
plt.show()
```

```python
# Feature Set-1
list_features=['Healthy life expectancy']
sns.boxplot(data=hd_2020.loc[:,list_features],
    notch=True, showcaps=False,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "green"})
plt.show()
```

#**Negara Paling Bahagia & Paling Tidak Bahagia**

```python
hd_2020_happiest_unhappiest=hd_2020[(hd_2020.loc[:,'Ladder score']>7.4)| (hd_2020.loc[:,'Ladder score']<3.5)]
sns.barplot(x='Ladder score', y='Country name',data=hd_2020_happiest_unhappiest,palette='Spectral',hue='Ladder score')
plt.show()
```

#**Negara Paling Dermawan dan Paling Tidak Dermawan**

```python
hd_2020=hd_2020[(hd_2020.loc[:,'Generosity']>0.2 )| (hd_2020.loc[:,'Generosity']<-0.2)]
sns.barplot(x='Generosity',y='Country name', data=hd_2020, palette='Spectral')
plt.title('Most Generous and Most Ungenerous Countries in 2020')
plt.show()
```
